```python
import numpy as np
import pandas as pd

sub = pd.read_csv('data/2015SampleSubmission.csv', index_col=0)
sub.tail()
```

```python
teamids = pd.read_csv('data/Teams.csv')
teamids.head()
```

```python
def getID(row):
    try:
        ID = teamids.loc[teamids['TeamName'] == row['Team'], 'TeamID'].values[0]
    except:
        ID = 0
    return ID
```

```python
teamids[np.array(['Dakota' in name for name in teamids['TeamName']])]
```

```python
playins15 = ['Manhattan', 'Hampton', 'BYU', 'Mississippi', 'Dayton', 'Boise St', 'North Florida', 'Robert Morris']
playins15 = pd.DataFrame(playins15, columns=['Team'])
playins15['TeamID'] = playins15.apply(getID, axis=1)
playins15
```

```python
playins15.to_csv('data/2015Playins.csv', encoding='ascii')
```

```python
regs15 = ['Kentucky', 'Playin1', 'Cincinnati', 'Purdue', 'West Virginia', 'Buffalo', 'Maryland', 'Valparaiso', 'Butler', 'Texas', 'Notre Dame', 'Northeastern', 'Wichita St', 'Indiana', 'Kansas', 'New Mexico St']
regs15 += ['Wisconsin', 'Coastal Car', 'Oregon', 'Oklahoma St', 'Arkansas', 'Wofford', 'North Carolina', 'Harvard', 'Xavier', 'Playin2', 'Baylor', 'Georgia St', 'VA Commonwealth', 'Ohio St', 'Arizona', 'TX Southern']
regs15 += ['Villanova', 'Lafayette', 'NC State', 'LSU', 'Northern Iowa', 'Wyoming', 'Louisville', 'UC Irvine', 'Providence', 'Playin3', 'Oklahoma', 'Albany NY', 'Michigan St', 'Georgia', 'Virginia', 'Belmont']
regs15 += ['Duke', 'Playin4', 'San Diego St', "St John's", 'Utah', 'SF Austin', 'Georgetown', 'E Washington', 'SMU', 'UCLA', 'Iowa St', 'UAB', 'Iowa', 'Davidson', 'Gonzaga', 'N Dakota St']
regs15 = pd.DataFrame(regs15, columns=['Team'])
regs15['TeamID'] = regs15.apply(getID, axis=1)
regs15
```

```python
regs15.to_csv('data/2015Regulars.csv', encoding='ascii')
```

```python
teams15 = pd.concat([regs15, playins15])
teams15 = teams15[np.array(['Playin' not in name for name in teams15['Team']])]
teams15 = teams15.reset_index(drop=True)
teams15.to_csv('data/2015Teams.csv', encoding='ascii')
```
